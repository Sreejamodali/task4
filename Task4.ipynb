```python
data_merge = pd.merge(df1,                 # Full outer join
                      df2,
                      on = ["Name","python","mysql","Previous Geekions","CodeKata Score","Department","Rising"],
                      how = "outer")
print(data_merge)
```

```python
data_merge.to_csv("data.csv")
```

```python
codekata_top=data_merge.loc[data_merge['CodeKata Score']>15000]
```

```python
codekata_top.shape
```

```python
# Exceeded expectations.csv
codekata_top.to_csv(" Exceeded expectations.csv")
```

```python
#Reached_expectations.csv
codekata_avg=data_merge.loc[(data_merge['CodeKata Score']<15000)&(data_merge["CodeKata Score"]>10000)]
codekata_avg.to_csv(" Reached_expectations.csv")
```

```python
#Needs_Improvement.csv
codekata_notbad=data_merge.loc[(data_merge['CodeKata Score']<10000)&(data_merge["CodeKata Score"]>7000)]
codekata_notbad.to_csv("Needs_Improvement.csv")
```

```python
#Unsatisfactory.csv
codekata_bad=data_merge.loc[data_merge['CodeKata Score']<7000]
codekata_bad.to_csv("Unsatisfactory.csv")
```

```python
import matplotlib.pyplot as plt
a=data_merge.Department.value_counts()['Computer Science and Engineering']
b=data_merge.Department.value_counts()['Electronics and Communication Engineering']
c=data_merge.Department.value_counts()['Electronics and Electrical Engineering']
sizes=[a,b,c]
lst=['Computer Science and Engineering','Electronics and Communication Engineering','Electronics and Electrical Engineering']
plt.figure(figsize=(2,2), dpi=200)
plt.pie(sizes, labels = lst,autopct='%1.0f%%', startangle=90)
plt.show()
```
